## Question 3. [Index] S&P 500 Market Corrections Analysis
Calculate the median duration (in days) of significant market corrections in the S&P 500 index.

For this task, define a correction as an event when a stock index goes down by more than 5% from the closest all-time high maximum.

Steps:

1. Download S&P 500 historical data (1950-present) using yfinance
2. Identify all-time high points (where price exceeds all previous prices)
3. For each pair of consecutive all-time highs, find the minimum price in between
4. Calculate drawdown percentages: (high - low) / high × 100
5. Filter for corrections with at least 5% drawdown
6. Calculate the duration in days for each correction period
7. Determine the 25th, 50th (median), and 75th percentiles for correction durations

 * Step 1: download S&P 500 data from 1950 - present

````python
sp500_ticker = yf.Ticker("^GSPC")
sp500_data = sp500_ticker.history(start=start, interval="1d")
sp500_data
````

* Step 2: Identify all-time high points (where price exceeds all previous prices)

````python
sp500_data['All_Time_High'] = sp500_data['Close'] > sp500_data['Close'].cummax().shift(1)
sp500_data
````

````python
# Filter only rows where an all-time high occurred

sp500_all_time_high = sp500_data[sp500_data['All_Time_High'] == True]
sp500_all_time_high
````

* Step 3: For each pair of consecutive all-time highs, find the minimum price in betweeen

````python
# Get all ATH dates
sp500_dates = sp500_all_time_high.index.to_list()

min_lows = []

# Loop through each pair of ATHs
for i in range(len(sp500_dates) - 1):
    start_date = sp500_dates[i]
    end_date = sp500_dates[i + 1]

    # Use the full DataFrame here!
    between_data = sp500_data.loc[start_date:end_date].iloc[1:-1]

    if not between_data.empty:
        min_close = between_data['Close'].min()
        min_close_date = between_data['Close'].idxmin()

        min_lows.append({
            'Start': start_date,
            'End': end_date,
            'min_close_date': min_close_date,
            'min_close_value': min_close
        })

# Final result
min_lows_df = pd.DataFrame(min_lows)
min_lows_df
````

* Step 4: Calculate drawdown percentages: (high - low) / high × 100

````python
# Add start high (all-time high before drawdown)
min_lows_df['start_high'] = min_lows_df['Start'].apply(lambda d: sp500_data.loc[d, 'Close'])

# Calculate drawdown percentage
min_lows_df['drawdown_pct'] = ((min_lows_df['start_high'] - min_lows_df['min_close_value']) /
                               min_lows_df['start_high']) * 100

# Optional: Round for readability
min_lows_df['drawdown_pct'] = min_lows_df['drawdown_pct'].round(2)

# View result
min_lows_df[['Start', 'End', 'min_close_date', 'min_close_value', 'start_high', 'drawdown_pct']].head()
````

* Step 5: Filter for corrections with at least 5% drawdown

````python
# Filter for drawdowns >= 5%
corrections_5pct = min_lows_df[min_lows_df['drawdown_pct'] >= 5]

# Display result
corrections_5pct[['Start', 'End', 'min_close_date', 'min_close_value', 'start_high', 'drawdown_pct']]
````

* Step 6: Calculate the duration in days for each correction period

````python
# Ensure datetime format
corrections_5pct['Start'] = pd.to_datetime(corrections_5pct['Start'])
corrections_5pct['End'] = pd.to_datetime(corrections_5pct['End'])

# Calculate duration
corrections_5pct['duration_days'] = (corrections_5pct['End'] - corrections_5pct['Start']).dt.days

# Display result
corrections_5pct[['Start', 'End', 'drawdown_pct', 'duration_days']].head()
````

* Step 7: Determine the 25th, 50th (median), and 75th percentiles for correction durations

````python
# Calculate percentiles
percentiles = corrections_5pct['duration_days'].quantile([0.25, 0.5, 0.75])

# Display result
print(percentiles)
````

* Context

````python
top_10_drawdowns = corrections_5pct.sort_values(by='drawdown_pct', ascending=False).head(10)
top_10_drawdowns[['Start', 'End', 'drawdown_pct', 'duration_days']]
````

* Calculate the median duration (in days) of significant market corrections in the S&P 500 index

````python
# Calculate the median duration of 5%+ corrections
median_duration = corrections_5pct['duration_days'].median()
print(f"Median duration of 5%+ corrections: {median_duration} days")
````

## Question 4. [Stocks] Earnings Surprise Analysis for Amazon (AMZN)
#### Calculate the median 2-day percentage change in stock prices following positive earnings surprises days.

Steps:

1. Load earnings data from CSV (ha1_Amazon.csv) containing earnings dates, EPS estimates, and actual EPS
2. Download complete historical price data using yfinance
3. Calculate 2-day percentage changes for all historical dates: for each sequence of 3 consecutive trading days (Day 1, Day 2, Day 3), compute the return as Close_Day3 / Close_Day1 - 1. (Assume Day 2 may correspond to the earnings announcement.)
4. Identify positive earnings surprises (where "actual EPS > estimated EPS" OR "Surprise (%)>0")
5. Calculate 2-day percentage changes following positive earnings surprises
6. Compare the median 2-day percentage change for positive surprises vs. all historical dates

Context: Earnings announcements, especially when they exceed analyst expectations, can significantly impact stock prices in the short term.

Reference: Yahoo Finance earnings calendar - https://finance.yahoo.com/calendar/earnings?symbol=AMZN

Additional: Is there a correlation between the magnitude of the earnings surprise and the stock price reaction? Does the market react differently to earnings surprises during bull vs. bear markets?)

* Step 1: Load the earnings data

````python
earnings_data = pd.read_csv('/content/ha1_Amazon.csv', sep=';')
earnings_data
````

* Step 2: Download complete historical price data using yfinance

````python
ticker_amzn = yf.Ticker('AMZN')
amzn_data = ticker_amzn.history(start='1997-01-01', interval='1d')
amzn_data
````

* Step 3: Calculate 2-day percentage changes for all historical dates: for each sequence of 3 consecutive trading days (Day 1, Day 2, Day 3), compute the return as Close_Day3 / Close_Day1 - 1. (Assume Day 2 may correspond to the earnings announcement.)

````python
# Compute 2-day return using Close_Day3 / Close_Day1 - 1
amzn_data['Close_t'] = amzn_data['Close']
amzn_data['Close_t_plus_2'] = amzn_data['Close'].shift(-2)

# Calculate 2-day return (Day 3 / Day 1 - 1)
amzn_data['2_day_return'] = (amzn_data['Close_t_plus_2'] / amzn_data['Close_t']) - 1

# Drop rows where shift created NaNs
amzn_2day_returns = amzn_data.dropna(subset=['2_day_return'])
amzn_2day_returns
````

Step 4: Identify positive earnings surprises (where "actual EPS > estimated EPS" OR "Surprise (%)>0")

````python
# Clean and convert numeric columns
earnings_data['EPS Estimate'] = pd.to_numeric(earnings_data['EPS Estimate'], errors='coerce')
earnings_data['Reported EPS'] = pd.to_numeric(earnings_data['Reported EPS'], errors='coerce')
earnings_data['Surprise (%)'] = pd.to_numeric(earnings_data['Surprise (%)'], errors='coerce')

# Filter positive surprises
positive_surprises = earnings_data[
    (earnings_data['Reported EPS'] > earnings_data['EPS Estimate']) |
    (earnings_data['Surprise (%)'] > 0)
]

positive_surprises = positive_surprises.reset_index(drop=True).dropna()
positive_surprises
````

Step 5: Calculate 2-day percentage changes following positive earnings surprises

````python
earnings_data['Earnings Date'] = pd.to_datetime(
    earnings_data['Earnings Date'].str.replace(r' [A-Z]{3}$', '', regex=True),
    format='mixed',
    errors='coerce'
).dt.tz_localize(None)
````

````python
amzn_data.index = amzn_data.index.tz_localize(None)
````

````python
# Filter positive surprises
positive_surprises = earnings_data[
    (earnings_data['Reported EPS'] > earnings_data['EPS Estimate']) |
    (earnings_data['Surprise (%)'] > 0)
].copy()

# Calculate 2-day returns after each positive earnings date
returns = []

for date in positive_surprises['Earnings Date']:
    # Find the nearest date in trading data (some earnings dates fall on non-trading days)
    if date not in amzn_data.index:
        date_idx = amzn_data.index.get_indexer([date], method='backfill')[0]
        date = amzn_data.index[date_idx]


    try:
        day1 = date
        day3 = amzn_data.index[amzn_data.index.get_loc(day1) + 2]  # 2 trading days later

        price_day1 = amzn_data.loc[day1]['Close']
        price_day3 = amzn_data.loc[day3]['Close']
        pct_return = (price_day3 / price_day1) - 1

        returns.append({'Earnings Date': day1, '2D Return': pct_return})
    except:
        # If day3 goes beyond available data or index error, skip
        continue

# Convert to DataFrame
returns_df = pd.DataFrame(returns)

# Merge with positive_surprises
result = pd.merge(positive_surprises, returns_df, on='Earnings Date', how='left')

result[['Earnings Date', 'EPS Estimate', 'Reported EPS', 'Surprise (%)', '2D Return']].dropna()
````

* Step 6: Compare the median 2-day percentage change for positive surprises vs. all historical dates

````python
# Shift index by 2 days to get the 2-day forward price
price_day1 = amzn_data['Close']
price_day3 = amzn_data['Close'].shift(-2)

# Calculate returns
amzn_2d_returns = (price_day3 / price_day1) - 1

amzn_2d_returns = amzn_2d_returns.dropna()
````

````python
# Median 2-day return after positive surprises
median_positive = result['2D Return'].median()
````

````python
# Median 2-day return for all days
median_all = amzn_2d_returns.median()
````

````python
print(f"Median 2-Day Return After Positive Surprises: {median_positive:.4%}")
print(f"Median 2-Day Return for All Historical Dates: {median_all:.4%}")
````
